```python
def str2Hump(text):
    """
    下划线转小驼峰 blood_detail -》 bloodDetail
    """ 
    arr = filter(None, text.lower().split('_'))
    res = ''
    j = 0
    for i in arr:
        if j == 0:
            res = i
        else:
            res = res + i[0].upper() + i[1:]
        j += 1
    return res
```

```python
x = str2Hump("blood_detail")
```

```python
def str2BigHump(text):
    """
    下划线转大驼峰 blood_detail -》 BloodDetail
    """ 
    arr = filter(None, text.lower().split('_'))
    res = ''
    j = 0
    for i in arr:
        print(1)
        print(res)
        if j == 0:
            res = i[0].upper() + i[1:]
        else:
            res = res + i[0].upper() + i[1:]
        j += 1
    return res
```
